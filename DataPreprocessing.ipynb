```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
from glob import glob
import random
import shutil
from tqdm import tqdm
```

```python
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
antigen_list=['ER','PR','HER2','Ki67']
save_train_path='../../data/IHC4BC_Compressed/train/'
save_test_path='../../data/IHC4BC_Compressed/test/'
for antigen in antigen_list:
    create_dir(f'{save_train_path}/{antigen}/HE')
    create_dir(f'{save_train_path}/{antigen}/IHC')
    create_dir(f'{save_test_path}/{antigen}/HE')
    create_dir(f'{save_test_path}/{antigen}/IHC')
    he_patient_list=glob(f'../../data/IHC4BC_Compressed/Images/HandE/{antigen}/**')
    ihc_patient_list=[f.replace('HandE','IHC') for f in he_patient_list]
    for i in tqdm(range(len(he_patient_list))):
        train_bool=random.random() < 0.9
        he_image_list=glob(f'{he_patient_list[i]}/**/*.jpg',recursive=True)
        ihc_image_list=glob(f'{ihc_patient_list[i]}/**/*.jpg',recursive=True)
        for j in range(len(he_image_list)):
            image_name=os.path.basename(he_image_list[j])
            shutil.copy(he_image_list[j],f'{save_train_path if train_bool else save_test_path}/{antigen}/HE/{image_name}')
            shutil.copy(ihc_image_list[j],f'{save_train_path if train_bool else save_test_path}/{antigen}/IHC/{image_name}')
    
```
